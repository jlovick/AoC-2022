```python
def create_piles(raw_data):
        last_line = raw_data[-1]
        piles = last_line.rstrip()[-1]
        print(piles)
        columns= {}
        for i in range(1,int(piles)+1):
            columns[i] = last_line.find(f"{i}")
        print(f" columns {columns}   in  {len(raw_data)} lines")
        output={}
        for col in range(1,len(columns)+1):
            output[col] = []
        print(output)
        for line in range(len(raw_data)-2,-1,-1):
            my_line = raw_data[line].rstrip()
            for col in range(1,len(columns)+1):
                print(f"{col} ---> {columns[col]} {len(my_line)}  :{my_line}")
                if columns[col] < len(my_line):
                    point_char = my_line[columns[col]]
                    if len(point_char.strip()) == 1: 
                        #print(f"found :{point_char}")
                        output[col].append(point_char)
                #print(output[col])
                    
        return output
```

```python
def setup_problem(input):
#    [D]    
#[N] [C]    
#[Z] [M] [P]
# 1   2   3 
    line_array = []
    with open(input) as file:
        for line in file:
            line = line.rstrip()
            if len(line) > 0:
                line_array.append(line)
            else:
                output = create_piles(line_array)
    return output
```

```python
def swap_items_pt1(instruction, piles_of_crates):

    for times in range(instruction["many"]):
        print(piles_of_crates)
        afrom = piles_of_crates[instruction["from"]].pop()
        print(f" got {afrom}")
        piles_of_crates[instruction["to"]].append(afrom)
        print(piles_of_crates)
    return piles_of_crates
```

```python
def swap_items_pt2(instruction, piles_of_crates):
    afrom=[]
    for times in range(instruction["many"]):
        print(piles_of_crates)
        afrom.append(piles_of_crates[instruction["from"]].pop())
        print(f" got {afrom}")
    afrom.reverse()
    piles_of_crates[instruction["to"]].extend(afrom)
    print(piles_of_crates)
    return piles_of_crates
```

```python
def read_instructions(input, piles_of_crates):
    line_array = []
    with open(input) as file:
        for line in file:
            line = line.rstrip()
            if line.find("from") > 0:
                instruction={}
                instruction["to"] = int(line[-1])
                instruction["from"] = int(line[-6])
                mv = line[4:7].rstrip()
                instruction["many"] = int(mv)
                print(instruction)
                piles_of_crates =  swap_items_pt2(instruction,piles_of_crates)
    return piles_of_crates
```

```python
#input = "./example/example_5th.txt"
input = "./puzzle_input/puzzle_input_5th.txt"
piles_of_crates = setup_problem(input)

piles_of_crates = read_instructions(input,piles_of_crates)
output = ""
for key in piles_of_crates.keys():
    if(len(piles_of_crates[key]) > 0):
        output += piles_of_crates[key][-1]
    
print(f" results {output}")
```
